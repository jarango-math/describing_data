```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

distribution_chisquare = np.random.chisquare(2, 100)
plt.hist(distribution_chisquare, color='c')
plt.axvline(distribution_chisquare.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distribution_chisquare.mean() + distribution_chisquare.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(distribution_chisquare.mean() - distribution_chisquare.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
distribution_geometric = np.random.geometric(.5, 100)
plt.hist(distribution_geometric, color='c')
plt.axvline(distribution_geometric.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distribution_geometric.mean()+distribution_geometric.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(distribution_geometric.mean()-distribution_geometric.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
distribution_exponential = np.random.exponential(2, 100)
plt.hist(distribution_exponential, color='c')
plt.axvline(distribution_exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distribution_exponential.mean()+distribution_exponential.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(distribution_exponential.mean()-distribution_exponential.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

Additionally:

    1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
    2. Add them together to create a third variable.

```python
normal_dist_1 = np.random.normal(5,0.5,100)
normal_dist_2 = np.random.normal(10,1,100)
normal_dist_added = normal_dist_1 + normal_dist_2
plt.hist(normal_dist_added, color='c')
plt.axvline(normal_dist_added.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal_dist_added.mean()+normal_dist_added.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normal_dist_added.mean()-normal_dist_added.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
